```kotlin
val input = """467..114..
               ...*......
               ..35..633.
               ......#...
               617....*43
               .....+..58
               -.592.....
               ......755.
               ...$.*....
               .664.598.."""

val lines = input.splitToSequence('\n').map { it.trim() }.toList()
val lineLen = lines[0].length
val grid = lines.map { it.toCharArray() }.toTypedArray()
val numRegex = "\\d+".toRegex()

val res = lines.mapIndexed { y, line ->
    line.toCharArray().foldIndexed(mutableListOf<Int>()) { x, res, char ->
        if (char == '*') {
            val top = grid.getOrNull(y - 1)?.slice(maxOf(0, x - 3)..minOf(lineLen - 1, x + 3))?.joinToString("") ?: ""
            val topMatches = numRegex.findAll(top).map { it.range }.filterNot { it.start > 4 || it.last < 2 }
                .map { top.slice(it).toInt() }.toList()
            var bottom =
                grid.getOrNull(y + 1)?.slice(maxOf(0, x - 3)..minOf(lineLen - 1, x + 3))?.joinToString("") ?: ""
            val bottomMatches = numRegex.findAll(bottom).map { it.range }.filterNot { it.start > 4 || it.last < 2 }
                .map { bottom.slice(it).toInt() }.toList()
            var left = grid.getOrNull(y)?.slice(maxOf(0, x - 3)..maxOf(0, x - 1))?.joinToString("")
            var right = grid.getOrNull(y)?.slice(minOf(lineLen - 1, x + 1)..minOf(lineLen - 1, x + 3))?.joinToString("")
            var leftAndRight = listOf(left, right).filter { it != null }.toList().joinToString(" ")
            val leftAndRightMatches =
                numRegex.findAll(leftAndRight).map { it.range }.map { leftAndRight.slice(it).toInt() }.toList()

            val allMatches = listOf(topMatches, bottomMatches, leftAndRightMatches).flatten()
            if (allMatches.size == 2) {
                res.add(allMatches.reduce(Int::times))
            }
        }
        res
    }
}.flatten().sum()

println(res)






```

```kotlin
val input = """467..114..
                ...*......
                ..35..633.
                ......#...
                617*......
                .....+.58.
                ..592.....
                ......755.
                ...$.*....
                .664.598..."""

fun <T> List<T>.uniques(): List<T> {
    return this.groupingBy { it }
        .eachCount()
        .filter { it.value == 1 }
        .keys
        .toList()
}

class GridLocation(val x: Int, val y: Int) {
     override fun equals(other: Any?): Boolean {
        if (this === other) return true
        if (other == null || this::class != other::class) return false

        other as GridLocation

        
        if (x != other.x) return false
        if (y != other.y) return false

        return true
    }
    
    override fun toString(): String {
        return "($x, $y)"
    }
    
    fun adjacentLocations(): List<GridLocation> {
        return listOf(
            GridLocation(x - 1, y - 1),
            GridLocation(x, y - 1),
            GridLocation(x + 1, y - 1),
            GridLocation(x - 1, y),
            GridLocation(x + 1, y),
            GridLocation(x - 1, y + 1), 
            GridLocation(x, y + 1), 
            GridLocation(x + 1, y + 1)
        )
    }
}


class Number {
    val digitsLocations = mutableListOf<GridLocation>()
    val digits = mutableListOf<Char>()
    fun pushDigit(x: Int, y: Int, digit: Char) {
        digitsLocations.add(GridLocation(x, y))
        digits.add(digit)
    }

    fun getValue() = digits.joinToString("").toInt()

    fun adjacentNeighborLocations(gridLocations: List<GridLocation>): kotlin.collections.Set<GridLocation> {
        var adjacentLocation = digitsLocations.flatMap {it.adjacentLocations()}.toSet()
        var intersection = adjacentLocation.intersect(gridLocations)
        return intersection
    }
}

//val symbols : MutableList<GridLocation> = mutableListOf()
//val numbers : MutableList<Number> = mutableListOf()
//var currentNumber: Number? = null

// Parse the input and 
var (numbers, symbols) = input.splitToSequence('\n')
    .map { it.trim() }
    .foldIndexed(Pair(mutableListOf<Number>(), mutableListOf<GridLocation>())) { y, (num, sym), line ->
        line.toCharArray().foldIndexed<Triple<MutableList<Number>, MutableList<GridLocation>, Number?>>(
            Triple(
                num,
                sym,
                null
            )
        ) { x, (numbers, symbols, number: Number?), c ->
            var currentNumber = number
            when {
                c.isDigit() -> {
                    val n = currentNumber ?: Number()
                    n.pushDigit(x, y, c)
                    currentNumber = n
                }

                c == '.' -> {
                    if (currentNumber != null) {
                        numbers.add(currentNumber)
                        currentNumber = null
                    }
                }

                else -> {
                    symbols.add(GridLocation(x, y))
                    if (currentNumber != null) {
                        numbers.add(currentNumber)
                        currentNumber = null
                    }
                }

            }
            if (x == line.length - 1 && currentNumber != null) {
                numbers.add(currentNumber)
            }
            Triple(numbers, symbols, currentNumber)
        }
        Pair(num, sym)
    }

println(symbols.toList())
//println(numbers.map { it.getValue() })

var res = numbers.filter { n -> n.adjacentNeighborLocations(symbols).size > 0}.map {it.getValue()}.sum()
println(res)
```
